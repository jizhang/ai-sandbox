```python
import os

import numpy as np
import pandas as pd
from openai import OpenAI

EMBEDDING_MODEL = 'text-embedding-3-small'

client = OpenAI(
    base_url=os.environ['OPENAI_API_BASE'],
    api_key=os.environ['OPENAI_API_KEY'],
)


def get_embeddings(texts):
    texts = [text.replace('\n', ' ') for text in texts]
    response = client.embeddings.create(
        input=texts,
        model=EMBEDDING_MODEL,
    )
    return [item.embedding for item in response.data]
```

```python
def cosine_similarity(vector_a, vector_b):
    dot_product = np.dot(vector_a, vector_b)
    norm_a = np.linalg.norm(vector_a)
    norm_b = np.linalg.norm(vector_b)
    epsilon = 1e-10
    return dot_product / (norm_a * norm_b + epsilon)


def search_product(df, query, n=3):
    embedding = get_embeddings([query])[0]
    df['similarity'] = df.embedding.apply(lambda x: cosine_similarity(x, embedding))
    return (
        df.sort_values('similarity', ascending=False)
        .head(n)
        .product_name
    )


search_product(df, '巨型背包')
```

```python
def recommend_product(df, product_name, n=3):
    embedding = df.loc[df.product_name == product_name, 'embedding'].values[0]
    df['similarity'] = df.embedding.apply(lambda x: cosine_similarity(x, embedding))
    return (
        df.sort_values('similarity', ascending=False)
        .head(n)
        .product_name
    )


recommend_product(df, '时尚休闲背包，容量大，出门必备，限时折扣！')
```

```python
# https://milvus.io/docs/build-rag-with-milvus.md
from pymilvus import MilvusClient

milvus_client = MilvusClient('milvus/milvus_demo.db')
collection_name = 'my_rag_collection'
if milvus_client.has_collection(collection_name):
    milvus_client.drop_collection(collection_name)

milvus_client.create_collection(
    collection_name=collection_name,
    dimension=len(df.iloc[0].embedding),
)
```

```python
data = []
for index, row in df.iterrows():
    data.append({
        'id': index,
        'vector': row.embedding,
        'text': row.product_name,
    })

milvus_client.insert(collection_name, data)
```

```python
def search_product_milvus(query, n=3):
    query_embedding = get_embeddings([query])[0]
    result = milvus_client.search(
        collection_name=collection_name, 
        data=[query_embedding], 
        limit=n, 
        output_fields=['text'],
    )
    return [item['entity']['text'] for item in result[0]]


search_product_milvus('巨型背包')
```
